<h1>Строим простую картограмму Pandas+Vincent </h1>

Сегодня же хотелось бы показать еще один способ отображения результатов анализа с помощью <a href="https://vincent.readthedocs.org/en/latest/index.html">Vincent</a>, который так же очень просто интегрируется с Pandas, хотя и займет это чуть больше действий, чем в случае с matplotlib. 

<h4>Введение</h4>

<b>Vincent</b> - это модуль предназначенный для трасляции данных из python в JavaScript библиотеки для визуализации <a href='http://d3js.org/'>D3js</a> и <a href='http://trifacta.github.io/vega/'>Vega</a>, которые в свою очередь дают большие возможности для интерактивной визуализации данных. 

Т.е. таким образом мы можем производить анализ на python, а графику к результатам мы сможем строить на js. Например это может быть удобно для визуализации каких-либо георафических данных и нанесения их на карту. Кроме того vincent обладает интеграцией с IPython Notebook, и, так же как и matplotlib, может выводить графику непосредственно в ней. 

В качестве демонстрации возможностей данного модуля, я предлагаю реализовать 2 задачи:
<ul>
    <li>Покажем динамику среднедушевого дохода населения и Центральному и Приволжскому федерельным округам</li>
    <li>Покажем на карте РФ распределение по субъектам РФ среднедушевого дохода за 2010</li></ul>
В качестве исходных данных возьмем статисткику <a href="http://www.gks.ru/bgd/regl/b12_14p/IssWWW.exe/Stg/d01/05-02.htm">сайта</a> Росстата. 

<h4>Анализ данных</h4>

Для начала давайте загрузим данные и посмортим нужна ли дополнительная обработка.

```python
import pandas as pd
import vincent

stat = pd.read_html('Data/AVGPeopleProfit.htm', header=0, index_col=0)[0]
```

Итак для загрузки данных в этот раз мы используем фнкцию <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.io.html.read_html.html?highlight=read_html#pandas.io.html.read_html">read_html()</a> (данная фукция появилась в pandas начиная с 0.12 версии). В качестве параметров в нашем случае передается 3 аргумента:<ol><li>Адрес html страницы</li><li>Номер строки, содержащей имена столбцов</li><li>Номер столбца, который будет использоваться в виде индекса</li></ol>

Хотелось бы отметить, что в отиличие от других фукнкций (например <i>read_csv</i>)  <i>read_html</i> возвращает не объект DataFrame, а список таких объектов. Это связано с тем, что страница может содержать не одну, а несколько таблиц и к нужной надо обращаться по номеру индекса.

После загрузки у нас получилась таблица следующего вида:

```python
stat.head()
```

Как можно заметить нужна будет небольшая обработка таблицы, т.к. в ней присутствует один столбец без названия и один столбец с пустыми значениями. Ну что же давайте выберем нужные нам столцы (со 2 по 13) остальные стобцы и поместим их в новый DataFrame.

```python
stat = stat[stat.columns[1:13]]
```

Теперь у нас получился набор данных, пригодный для работы. Конечно режет глаз названия столбцов, но такие имена нам ничуть не помешают решить поставленные задачи.

```python
stat.head()
```

Итак, давайте приступим к выполнениию первой задачи по визуализации данных 2 округов. Чтобы получить данные, на основе которых мы построим график, необходимо выбрать интересующие нас округа (Московский и Приволжский), а затем транспонировать полученную таблицу. Сделать это можно так:

```python
fo = [u'Приволжский федеральный округ',u'Центральный федеральный округ']
fostat = stat[stat.index.isin(fo)].transpose()
```

В приведенном выше коде, мы сначала фильтруем наш набор данных по нужным нам округам при момощи функции <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.isin.html#pandas.Series.isin">isin()</a>, которая проверяет значение столбца в заданее заданном списке (аналог оператора IN в SQL). Затем используем функцию <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.transpose.html?highlight=transpose#pandas.DataFrame.transpose">transpose()</a> для транспонирования полученного набор данных и записываем результат в новый DataFrame.

Полученная таблица выглядит так:

```python
fostat
```

Как можно заметить названия индексы в таблице у нас теперь равны номеру года в числовом формате. Это не очень удобно, поэтому давайте поменяем индекс на формат даты:

```python
fostat.set_index(pd.date_range('1999','2011', freq='AS'), inplace=True)
```

Фукция <a href="http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.set_index.html?highlight=set_index#pandas.DataFrame.set_index">set_index()</a> используется для задания нового индекса в DataFrame. В нашем случае ей передается 2 параметра:<ol><li>Список новых значений индекса(может быть также названием столбца)</li><li>Парамерт означает, что мы заменяем индекс в текущем наборе, если он будет False индекс не сохранится<ol>

Теперь данные у нас полностью готовы для построения графика. Итак если вы работаете в IPython Notebook и хотите видеть результат в реальном времени то для интеграции необходимо вызвать функцию <i>initialize_notebook()</i>. Выглядеть это будет следующим образом:

```python
vincent.core.initialize_notebook()
```

Теперь нам нужно создать объект соответствующий типу диаграммы (полный список объектов можно увидеть в <a href="https://vincent.readthedocs.org/en/latest/">документации</a>). В нашем случае это будут линейный график. Код будет следующий:

```python
line = vincent.Line(fostat) #создаем объект графика
line.axis_titles(x=u'Год', y=u'тыс. руб') #задаем названия осей
line.legend(title=u'ЦФО vs ПФО') #выводим легенду и задаем ей заголовок
```

Вывести график можно с помощью функции <i>display()</i>:

```python
line.display()
```

<h4>Построение картограммы</h4>

Ну что же с первой задачей мы справились. Теперь давайте перейдем ко второй. Для ее решения нам понадобится TopoJSON файл с картой РФ, а также справочник регионов. Подробно о том как их получить и что это такое можно почитать <a href="http://habrahabr.ru/post/181766/">здесь</a>. Для начала давайте загрузим справочник регионов с помощью <i>read_csv</i>, описанной в одной из <a href="http://habrahabr.ru/post/196980/">предыдущих</a> статей:

```python
spr = pd.read_csv('Data/russia-region-names.tsv','\t', index_col=0, header=None, names = ['name','code'], encoding='utf-8')
```

Как можно заметить тут появилось несколько дополнительный параметров:<ul><li><i>index_col</i> - задает номер столбца который будет использоватся в качестве индекса</li><li><i>header</i> - в нашем случае означает, что для определения заголовков мы не используем строки из файла</li><li><i>names</i> - получает список, элементы которого будут названиями столбцов</li><li><i>encoding</i> - задает кодировку в котрой хранится файл</li></ul>

Если мы посмотрим внимательно, на наш набор данных <i>stat</i>, то можно увидеть, что его некоторые элементы содержат сноски тип '1)' и '2)', которые при парсинге с помощью read_html() перекодировались в обычные символы и добавились в конце соответствующих строк в индексом сотлбце. Кроме того перед именами городов в нашем наборе строит буква 'г. ', а в справочнике ее нет. Все эти мелочи влияют на то, что когда мы будем объединять набор со стат. данными и справочник, чтобы подтянуть коды к регионам, у нас будут регионы без кода.

Исправить это можно следующим образом:

```python
new_index = stat.index.to_series()
new_index = new_index.str.replace(u'(2\))|(1\))|(г. )','')
```

Первая строка означает, что мы выделяем индексный столцеб в отдельную новую серию. Во втрой строке мы заменяем значения соответствующие регулярному выражению на пустые.

Теперь нам надо значения индекса заменить значениями из нового нвбора. Как показовалось выше, сделать это можно так:

```python
stat.set_index(new_index, inplace=True)
```

Теперь мы можем объединить наш набор данных со справочником, чтобы получить коды регионов

```python
RegionProfit = stat.join(spr, how='inner')
```

Наши данные после всех манипуляций выглядят так:

```python
RegionProfit.head()
```

Итак перейдем к непосредственному построению карты и нанесению на нее данных. Для начача нам нужно создать словарь с описание нашей карты:

```python
geo_data = [{'name': 'rus', #имя карты
             'url': 'RusMap/russia.json', #путь до TopoJSON файла с картой
             'feature': 'russia'}] #имя объекта из файла карты
```

Теперь давайте создадим объект нашей карты и привяжем к нему наши данные. Сделать это можно функцией Map():

```python
vis = vincent.Map(data=RegionProfit, geo_data=geo_data,scale=700, projection='conicEqualArea', rotate = [-105,0], center = [-10, 65],
          data_bind=2011, data_key='code',
          map_key={'rus': 'properties.region'})
```

В качестве параметров функция принимает следующие аргументы:<ul><li><i>data</i> - набор с данными</li><li><i>geo_data</i> - объект с нашей картой</li><li><i>projection</i>,- проекция в которой наша карта будет отображатся</li><li><i>rotate, center, scale</i> - <a href='http://habrahabr.ru/post/181766#prj_param'>параметры проекции</a></li><li><i>data_bind</i> - столбец с данными, которые будут отображатся</li><li><i>data_key</i> - поле с кодом по которому будет осуществлятся связка карты и данных</li><li><i>map_key</i> - словарь тип <i><b>{'имя объекта с картой':'имя свойства по которому осуществляется привязка'}</b></i></li><ul>


Но тут наш ожидаем неожиданность: в авторской версии vincent, параметр <b>rotate</b> может быть только целочисленный. Для корретного отображения нашей карты, нужна возможность, чтобы данный параметр смог принимать значения в виде списка. Чтобы это исправить идем в файл <i>%PYTHON_PATH%\lib\site-packages\vincent\transforms.py</i> заменям кусок кода отвечающий за проверку типа переменных:
<code>
    @grammar(int)
    def rotate(value):
        """The rotation of the projection"""
        if value < 0:
            raise ValueError('The rotation cannot be negative.')
</code>
на :
<code>
    @grammar(list)
    def rotate(value):
		if len(value) != 2:
			raise ValueError('len(center) must = 2')
</code>	

Теперь наш объект создастся корретно. Осталось поднастроить наш объект. Для начала давайте сделаем наши границы между меджу объектами менее заметными. Для этого у нас существуют метки (marks), которые являются основными строительным элементом. Подробней про них написано в документации к <a href="https://github.com/trifacta/vega/wiki/Marks">Vega</a>. В нашем случае код выглядит так:

```python
vis.marks[0].properties.enter.stroke_opacity = vincent.ValueRef(value=0.5)
```

Теперь давайте зададим чтобы наши значения в зависимости от группы окрашивались в разные цвета. Сделать это можно с помощью объектов <a href="https://github.com/trifacta/vega/wiki/Scales">Scales</a>,  предназначеных для перевода значений данных (числовых, строковых, дат и тд) в значения для отображения (писксели, цвета, размеры)

```python
vis.scales['color'].type = 'threshold' #задает тип шкалы
vis.scales['color'].domain = [10000, 15000, 20000, 25000, 30000] #задаем набор значений данных для группировки
vis.legend(title=u'Доходы руб.') #вводим легенду карты
```

Ну что же карта настроена, теперь ее можно посмотреть что у нас получилось. Как указаволось выше для этого можно использовать функцию <i>display()</i>, но у меня по неизвестным причинам она не сработала, поэтому я сначала выгрузил ее в итоговый json файл с помощью функции <i>to_json()</i>:

```python
vis.to_json('example_map.json', html_out=True, html_path='example_map.html')
```

В качестве параметров ей передается 3 параметра: <ol><li>имя итогового файла</li><li>html_out указывает что необходимо создать еще и html файл оболочки</li><li>html_path - задает путь до html файла</li></ol>

Чтобы посмотреть наш html файл <a href="https://github.com/wrobstory/vincent/tree/master/examples">нужен</a> простой HTTP сервер входящий в Python. Для его запуска в командной строке надо выполнить команду: 

<code>python -m SimpleHTTPServer 8000</code>
