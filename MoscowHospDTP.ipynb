#### Загрузка и первичная обработка данных

Загрузим данные и посмотрим на них:

```python
from pandas import read_csv
from statsmodels.api import tsa, graphics, stats
from pandas.tseries.tools import to_datetime
```

```python
dataset = read_csv('http://hubofdata.ru/storage/f/2013-08-18T16%3A58%3A34.789Z/timeseries-hosp.csv',
                   '|',
                   index_col = ['thedate'],
                   usecols = ['thedate','hosp','hosp_children'],
                   parse_dates = ['thedate'],                
                   date_parser=to_datetime,
                   dayfirst=True
                   )
dataset.head()
```

Итак, как можно заметить функция <a href='http://pandas.pydata.org/pandas-docs/stable/generated/pandas.io.parsers.read_csv.html?highlight=read_csv#pandas.io.parsers.read_csv'>read_csv()</a>, в данном случем помимо указания параметров, которые задают используемые колонки и индекс, можно заметить еще 3 параметра для работы с датой. Остановимся на них поподробнее. 

<b>parse_dates</b> задает имена столбцов, которые будут преобразованы в тип <i>DateTime</i>. Стоит отметить, что если в данном столбце будут пустые значения парсинг не удастся и вернется столбец типа <i>object</i>. Чтобы этого избежать надо добавить параметр <i><b>keep_default_na</b>=False</i>.

Следующий параметр <b>date_parser</b> задает функцию для парсинга, по умолчанию его значение равно <i>dateutil.parser.parser</i>. В нашем случае, мы воспользуемся встроенной в pandas функцией <a href='http://pandas.pydata.org/pandas-docs/stable/generated/pandas.tseries.tools.to_datetime.html?highlight=to_datetime#pandas.tseries.tools.to_datetime'>to_datetime()</a>, которая переводит строку в тип <i>DateTime</i>.

Заключительный параметр <b>dayfirst</b> указывает функции парсинга, что первое в строке первым идет день, а не наоборот. Если не задать этот параметр, то функция может не правильно преобразовывать даты и путать месяц и день местами. Например <i>01.02.2013</i> будет преобразовано в <i>02-01-2013</i>, что будет неправильно.

Также стоит отметить, что в нашем загруженном DataFrame'e индекс будет временной последовательность, что очень удобно для анализа (это будет показано в следующем разделе).

Данные загружены и обработаны и можно перейти к анализу данных.

#### Анализ данных

Итак, первое что мы можем посмотреть из нашего набора, это общее количество госпитализированных по месяцам c разбиением на детей и взрослых. 

Первое, что нам надо сделать, это изменить значение индекса, для того, чтобы в нем содержались даты конца месяца. Один из способов сделать это - использовать свойство <a href='http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html?highlight=resample#pandas.DataFrame.resample'>resample()</a>. Данное свойство преобразовывает индекс в последовательность меньшей размероности и использует заданную функцию для аггрегирования значений. Выглядит это так:

```python
stat_month = dataset.resample('M', how='sum')
```

Теперь можно посмотрить график который у нас получиться:

```python
stat_month.plot(figsize=(15,4))
```

Как можно заметить из графика, количество госпитализированных постепенно уменьшается, в то время как количество детей остается приблизтельно на одном уровне. Поэтому дальнейший анализ я предлагаю проводить, только по количеству госпитализированных взрослых.

Для начала проверим ряд на стационарность с помощью обобщенного теста <a href='http://ru.wikipedia.org/wiki/%D0%A2%D0%B5%D1%81%D1%82_%D0%94%D0%B8%D0%BA%D0%B8_%E2%80%94_%D0%A4%D1%83%D0%BB%D0%BB%D0%B5%D1%80%D0%B0'>Дикки-Фуллера</a>. Это очень легко сделать с помощью пакета <a href='http://statsmodels.sourceforge.net/stable/index.html'>statmodels</a>, который включает большой набор средств и методов для работы с временными рядами и не только. 

```python
print 'adf: ', tsa.adfuller(stat_month.hosp.dropna())[0] 
print'Critical values: ', tsa.adfuller(stat_month.hosp.dropna())[4]
```

Как можно заметить, при уровне значимости в 10% полученное значение статистики меньше критического значения, что может свидетельствовать о стационарности ряда.

В качестве замечания, надо отметить, что если нам нужен будет уровень точности в 5%, то необходимо будет перейти к <i>разности первого порядка</i>. Так назвается разность текущего значения с предыдущим ($\Delta^1_t = Y_t — Y_{t-1}$). В нашем случае переход к разности 1-ого порядка выглядел бы так:

```python
delta = stat_month.hosp - stat_month.hosp.shift()
```

В коде выше функция <a href='http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.shift.html?highlight=shift#pandas.Series.shift'>shift()</a> выполняет смещение данного ряда на нужный промежуток.

Теперь для расчета наличия автокорреляции расчитаем значение критерия <a href='http://ru.wikipedia.org/wiki/%CA%F0%E8%F2%E5%F0%E8%E9_%C4%E0%F0%E1%E8%ED%E0_%97_%D3%EE%F2%F1%EE%ED%E0'>Дарбина-Уотсона</a>, он считается по формуле:

$$d ={\sum_{t=1}^{n}(y_{t+1}-y_t)^2}/{\sum_{t=1}^{n}y_t^2}$$

где $y_{(t+1)}$ и $y_t$ – последовательные значения временного ряда

Значение статистики Дарбина-Уотсона изменяется в диапазоне от 0 до 4. При этом d = 2 указывает на отсутствие автокорреляции элементов временного ряда. Если d меньше двух, то имеет место положительная автокорреляции, а больше двух - отрицательная.

```python
ts = stat_month.hosp.dropna() #уберем все пустые значения из нашего ряда
stats.durbin_watson(ts)
```

В нашем случает данный критерий показывает на положительную автокорреляцию первого порядка. К недостаткам данного критерия относится, то что он не применим к моделям авторегресии о которых речь пойдет далее, а также не способен выявлять автокорреляцию второго и более высоких порядков. Учитывая вышесказанное, надо отметить, что его вычесление приведено для ознакомления.

Исходя из нашего предположения о стационарности ряда, мы можем построить <a href='http://ru.wikipedia.org/wiki/ARMA'>модель ARMA</a>. Данная модель, это комбинация моделей <a href='http://ru.wikipedia.org/wiki/%D0%90%D0%B2%D1%82%D0%BE%D1%80%D0%B5%D0%B3%D1%80%D0%B5%D1%81%D1%81%D0%B8%D0%BE%D0%BD%D0%BD%D0%B0%D1%8F_%D0%BC%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C'>авторегрессии(AR)</a> и <a href='http://ru.wikipedia.org/wiki/%D0%9C%D0%BE%D0%B4%D0%B5%D0%BB%D1%8C_%D1%81%D0%BA%D0%BE%D0%BB%D1%8C%D0%B7%D1%8F%D1%89%D0%B5%D0%B3%D0%BE_%D1%81%D1%80%D0%B5%D0%B4%D0%BD%D0%B5%D0%B3%D0%BE'>скользящего среднего(MA)</a>.

<b>ARMA(p, q)</b>, имеет следующий вид:

$X_t = \alpha_0 + \alpha_1X_{t-1}+ \alpha_2X_{t-2} + \ldots+ \alpha_pX_{t-p} + ε_t – \beta_1ε_{t-1}– \beta_2ε_{t-2}–\ldots–\beta_qε_{t-q}$

, где:

$ε_t$ – процесс типа <a href="http://ru.wikipedia.org/wiki/%D0%91%D0%B5%D0%BB%D1%8B%D0%B9_%D1%88%D1%83%D0%BC">«белый шум»</a>,

$\alpha_1, \ldots, \alpha_p$ — авторегрессионные коэффициенты, 

$\beta_1, \ldots, \beta_q$ — коэффициенты скользящего среднего

Для начала построим определим размерности <i>p</i> и <i>q</i> нашей модели. Для этого нам  нужно построить коррелограммы автокорреляционной функции (ACF) и частично автокорреляционной функции (PACF), т.к. <i>ACF</i> поможет нам определить порядок q, а <i>PACF</i> - порядк p.

Порядок <b>q</b> будет соотвтетсвовать номеру лага, после которого значение функции <i>ACF</i> практически равно 0.

Соответственно порядок <b>p</b> будет определять номер лага, после которого функция <i>PACF</i> будет практически равна 0.

Создатели пакета <b>statsmodels</b>, позаботились об этом, включив функции <a href='http://statsmodels.sourceforge.net/stable/generated/statsmodels.graphics.tsaplots.plot_acf.html'>plot_acf()</a> и <a href='http://statsmodels.sourceforge.net/stable/generated/statsmodels.graphics.tsaplots.plot_pacf.html'>plot_pacf()</a> в свой модуль. Данные функции выводят графики <i>ACF</i> и <i>PACF</i>, у которых по оси X откладываются номера лагов, а по оси Y значения соответствующих функций.

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
fig = graphics.tsa.plot_acf(ts, lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = graphics.tsa.plot_pacf(ts, lags=40, ax=ax2)
```

Как можно заметить на графиках есть значения функций выпадающие за $2\pm\sigma$, поэтому, в <b>p</b> и <b>q</b>, возьмем номера лагов в которых это происходит.

#### Построение модели и прогнозирование

Ну что же давайте теперь построим собственно модель:

```python
arma_model = tsa.ARMA(ts, (1,4), freq='M').fit()
```

Итак, у нас есть модель, давайте теперь попробуем на спрогнозировать количество госпитализированных на 2014 год. Сделать это можно с помощью функции <a href='http://statsmodels.sourceforge.net/stable/generated/statsmodels.tsa.arima_model.ARMA.predict.html#statsmodels.tsa.arima_model.ARMA.predict'>predict()</a>:

```python
pred = arma_model.predict('2013', '2014', dynamic=True)
```

Посмотрим информацию по нашей модели:

```python
print arma_model.summary()
```

В форме вывода помимо коэффициентов и их весов мы также можем заметить параметры используемые для оценки моделей, такие как:
<ul>
<li><a href='http://ru.wikipedia.org/wiki/%D0%98%D0%BD%D1%84%D0%BE%D1%80%D0%BC%D0%B0%D1%86%D0%B8%D0%BE%D0%BD%D0%BD%D1%8B%D0%B9_%D0%BA%D1%80%D0%B8%D1%82%D0%B5%D1%80%D0%B8%D0%B9_%D0%90%D0%BA%D0%B0%D0%B8%D0%BA%D0%B5'>Информационный критерий Акаике</a> (AIC)</li>

</ul>

В заключении построим график с нашим прогнозом и выведем его формулу:

```python
ax = stat_month.hosp.plot(figsize=(12,6))
ax = pred.plot(ax=ax, style='r--', label='Prediction hosp');
ax.legend();
```

Конечный вид нашей модели:

$$\Delta ts_t = 862.9021 + 0.8950*\Delta ts_{t-1} + 0.0011*\varepsilon_{t-1}+0.2107*\varepsilon_{t-2}-0.1153*\varepsilon_{t-3}-0.0119*\varepsilon_{t-4}$$
