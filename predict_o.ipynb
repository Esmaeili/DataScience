```python
%load_ext rmagic
from pandas import read_csv, DataFrame, Series
import statsmodels.api as sm
import rpy2.robjects as R
from rpy2.robjects.packages import importr
import pandas.rpy.common as com
from pandas import date_range

dataset = read_csv('DataSets/tovar_moving.csv',';', index_col=['date'], parse_dates=['date'], dayfirst=True)
otg = dataset.qty
```

Перейдем к недельным данным:

```python
w = otg.resample('w', how='sum')
w.plot(figsize=(12,6), color='red')
```

Преобразуем полученные данные во временную последовательность Python:

Из графика заметна сезонность в 12 мес. Поэтому для ADF теста возьмем 52 лагов (т.к. это 12 мес = 52 недели).

```python
w_log = log(w)
```

Набор данных имеет нормальное распределение.

С помощью теста Дикки-Фулера проверяем гипотезу о стационароности ряда c помощью R.

```python
stats = importr('stats')
tseries = importr('tseries')

r_df = com.convert_to_r_dataframe(DataFrame(w_log))
y = stats.ts(r_df)
ad = tseries.adf_test(y, alternative="stationary", k=52)
a = ad.names[:5]
{ad.names[i]:ad[i][0] for i in xrange(len(a))}
```

Исходный ряд не стационарен. Проверяем гипотезу о стационароности ряда первых разностей c помощью библиотек R:

Первые разности с помощью Python:

```python
diff1lev = w.diff(periods=1).dropna()
```

Тест Дикки-Фулера:

```python
print 'p.value: %f' % sm.tsa.adfuller(diff1lev, maxlag=52)[1]
```

```python
diff1lev.plot(figsize=(12,6))
```

Ряд первых разностей стационарен. Тренд отсутствует.

Т.к. на наших данных присутствует сезонность. То нужно от нее избавиться. Поэтому возьмем сезонную разность ряда первых разностей и проверим ее на стационарность, на этот раз сделаем это с помощью R:

```python
diff1lev_season = diff1lev.diff(52).dropna()
```

```python
r_df = com.convert_to_r_dataframe(DataFrame(diff1lev_season))
y = stats.ts(r_df)
ad = tseries.adf_test(y, alternative="stationary", k=52)
a = ad.names[:5]
{ad.names[i]:ad[i][0] for i in xrange(len(a))}
```

Берем первые разности сезонности:

```python
diff1lev_season1lev = diff1lev_season.diff().dropna()
print 'p.value: %f' % sm.tsa.adfuller(diff1lev_season1lev, maxlag=52)[1]
```

Ряд стационарен. Определены параметры для модели SARIMA: d=1, D=1

Построим коррелограммы acf и pacf для определения порядков MA и AR, а также для определения параметров сезонности:

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff1lev_season1lev.values.squeeze(), lags=150, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff1lev_season1lev, lags=150, ax=ax2)
```

Для исходного ряда параметры SARIMA q = 1, p = 1, Q = 1, P = 1

Получается модель SARIMA(4,1,13)(0,1,0)12

```python
r_df = com.convert_to_r_dataframe(DataFrame(w))
y = stats.ts(r_df)
order = R.IntVector((4,1,13))
season = R.ListVector({'order': R.IntVector((0,1,0)), 'period' : 52})
model = stats.arima(y, order = order, seasonal=season)
```

Делаем прогноз

Способ 1:

```python
%Rpush model
%R plot(forecast(model))
```

Способ 2:

```python
forecast = importr('forecast')
f = forecast.forecast(model)
```

Переводим результат в тип Python

```python
pred = [i[1] for i in f[3].iteritems()]
dt = date_range(w.index[-1], periods=len(pred)+1,freq='W')[1:]
pr = Series(pred, index = dt)
```

Рисуем проноз

```python
w.plot()
pr.plot(color = 'red')
```
