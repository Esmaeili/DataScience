Теперь же загрузим, нужные библиотеки:

```python
from pandas import read_csv, DataFrame, Series
import statsmodels.api as sm
import rpy2.robjects as R
from rpy2.robjects.packages import importr
import pandas.rpy.common as com
from pandas import date_range
```

Теперь, как и в предыдущей статье, загрузим данные и перейдем к недельным интервалам:

```python
dataset = read_csv('DataSets/tovar_moving.csv',';', index_col=['date'], parse_dates=['date'], dayfirst=True)
otg = dataset.qty
w = otg.resample('w', how='sum')
w.plot(figsize=(12,6))
```

Итак, из графика можно заметить ежегодную сезонность (52 недели) и ярко выраженный тренд. Поэтому перед построением модели нам необходимо избавиться от тренда и сезонности.

#### Предварительный анализ данных

Итак для начала, прологорифмируем исходный ряд, для выравнивания значений: 

```python
w_log = log(w)
w_log.plot(figsize=(12,6))
```

Как видно у нас в графике присутствует сезонность и соответственно ряд не стационарен. Проверим это с помощью теста Дикки-Фулера, который проверяет гипотизу о наличии единичных корней и соотвтвенно если они есть ряд будет не стационарным. Как провести данный тест с помощью библиотеки <b>statsmodels</b>, я показывал в прошлый раз. Сейчас я продемонстрирую как это можно сделать с помощью функции <a href="http://www.inside-r.org/packages/cran/tseries/docs/adf.test">adf.test()</a> из <b>R</b>.

Итак, данная функция находится в R-ой библиотеке <a href="http://cran.r-project.org/web/packages/tseries/index.html">tseries</a>. Она предназначена для анализа временных рядов и устанавливается дополнительно. Загрузить нужную библиотеку можно с помощью функции <a href="http://rpy.sourceforge.net/rpy2/doc-dev/html/robjects_rpackages.html?highlight=importr#rpy2.robjects.packages.importr">importr()</a>.

```python
stats = importr('stats')
tseries = importr('tseries')
```

Можно заметить, что кроме <i>tseries</i>, мы загрузили еще и библиотеку <i>stats</i>. Она нам понадобитьсядля преобразования типов.

Теперь необходимо перевести данные из типа Python в тип понятный R. Сделать это можно с помощью функции <a href="http://pandas.pydata.org/pandas-docs/stable/r_interface.html">convert_to_r_dataframe()</a> на вход которой подается DataFrame, а на выходе получается вектор для R

```python
r_df = com.convert_to_r_dataframe(DataFrame(w_log))
```

Итак, вектор есть следующим шагом надо перевести его в формат временного ряда. Для этого в R существует функция <a href="http://stat.ethz.ch/R-manual/R-patched/library/stats/html/ts.html">ts()</a>, вызов ее будет выглядеть так:

```python
y = stats.ts(r_df)
```

Предварительная подготовка данных закончена и мы можем вызвать нужную нам функцию:

```python
ad = tseries.adf_test(y, alternative="stationary", k=52)
```

В качестве параметров ей передается временный ряд и кол-во лагов, для которых будет расчитываться тест. В экономических моделях принято брать данное значение равное году, а т.к. данные у нас еженедельные, а в году 52 недели, то данный показатель 52.

Теперь в переменной <i>ad</i> содержится R-объект. Его структура описана в виде списка, описание которого мне найти не удалось. Поэтому с помощью визуального анализа я написал код, который выводит результат работы функции в понятном виде:

```python
a = ad.names[:5]
{ad.names[i]:ad[i][0] for i in xrange(len(a))}
```

Исходя из результатов теста, на исходный ряд не стационарен. Соответственно теперь проверим на стационарность ряд первых разностей.

Для начала получим первые разности с помощью Python, а затем применим ADF-тест:

```python
diff1lev = w.diff(periods=1).dropna()
print 'p.value: %f' % sm.tsa.adfuller(diff1lev, maxlag=52)[1]
```

```python
diff1lev.plot(figsize=(12,6))
```

Ряд первых разностей, по итогам теста, оказался стационарным. А график помогает нам убедиться, что тенденция отсутствует. Осталось избавиться от сезонности.

Для этого нужно взять сезонную разность, от нашего получившегося ряда. Подробнее можно прочитать <a href="https://onlinecourses.science.psu.edu/stat510/?q=node/67">тут</a>. Если полученный ряд будет стационармы необходимо будет вязть его превую разность и проверить ее.

Возьмем сезонную разность:

```python
diff1lev_season = diff1lev.diff(52).dropna()
```

Проверим ее на стационарность с помощью ADF-тест из R:

```python
r_df = com.convert_to_r_dataframe(DataFrame(diff1lev_season))
y = stats.ts(r_df)
ad = tseries.adf_test(y, alternative="stationary", k=52)
a = ad.names[:5]
{ad.names[i]:ad[i][0] for i in xrange(len(a))}
```

Итак, ряд не стационарен, возьмем его первые разности:

```python
diff1lev_season1lev = diff1lev_season.diff().dropna()
print 'p.value: %f' % sm.tsa.adfuller(diff1lev_season1lev, maxlag=52)[1]
```

Получившийся ряд стационарен. Теперь можно перейти к построению модели 

#### Построение модели.

Построим коррелограммы acf и pacf для определения порядков MA и AR, а также для определения параметров сезонности:

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff1lev_season1lev.values.squeeze(), lags=150, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff1lev_season1lev, lags=150, ax=ax2)
```

Для исходного ряда параметры SARIMA q = 1, p = 1, Q = 1, P = 1

Получается модель SARIMA(4,1,13)(0,1,0)12

```python
r_df = com.convert_to_r_dataframe(DataFrame(w))
y = stats.ts(r_df)
order = R.IntVector((4,1,13))
season = R.ListVector({'order': R.IntVector((0,1,0)), 'period' : 52})
model = stats.arima(y, order = order, seasonal=season)
```

Делаем прогноз

```python
forecast = importr('forecast')
f = forecast.forecast(model)
```

Способ 1:

```python
%Rpush model
%R plot(forecast(model))
```

Способ 2:

Переводим результат в тип Python

```python
pred = [i[1] for i in f[3].iteritems()]
dt = date_range(w.index[-1], periods=len(pred)+1,freq='W')[1:]
pr = Series(pred, index = dt)
```

Рисуем проноз

```python
w.plot()
pr.plot(color = 'red')
```
