```python
from pandas import read_csv, DataFrame
```

```python
def distr_method_pie(history):
    plt.figure(figsize=(7,7))
    method_stats = history[history/float(history.sum()) > .02]
    colors=['yellow','green','red','blue', 'lightskyblue', 'lightcoral','gray', 'white']

    max_part = [0.1] + [0 for i in xrange(method_stats.__len__()-1)]

    plt.pie(method_stats, 
            labels=method_stats.index.tolist(), 
            explode=tuple(max_part),
            colors=colors,
            autopct='%1.1f%%', 
            shadow=True)
    plt.show()
```

```python
def dynamic_method_graph(method, history, diap):
    top_method = history[:4].index
    tmp = method[method.METHOD_DESC.isin(top_method)]
    top_method_grop = tmp.groupby(['DATE','METHOD_DESC']).count().METHOD.reset_index()
    pivot_method = top_method_grop.pivot('DATE','METHOD_DESC','METHOD').fillna(0)
    pivot_method.resample(diap,'sum').dropna().plot(figsize=(15,5))
```

```python
stats = read_csv('..\DataSets\ufc\statistics_ufc.csv', ';', parse_dates=['DATE'])
```

```python
all_stat = stats.groupby('METHOD').count().METHOD_DESC
all_stat = all_stat[all_stat > 0]
```

```python
all_stat
```

```python
plt.figure(figsize=(7,7))
plt.pie(all_stat, 
        labels=all_stat.index.tolist(), 
        autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
submitions = stats[stats.METHOD == 'SUB']
all_history_sub = submitions.groupby('METHOD_DESC').count().FIGHTER_WIN.order(ascending=False)
ko = stats[stats.METHOD == 'KO/TKO']
all_history_ko = ko.groupby('METHOD_DESC').count().FIGHTER_WIN.order(ascending=False)
```

```python
distr_method_pie(all_history_sub)
```

```python
dynamic_method_graph(submitions, all_history_sub,'M')
savefig('submintion_year_dynamic.png')
```

```python
dynamic_method_graph(submitions, all_history_sub,'AS')
savefig('submintion_month_dynamic.png')
```

```python
distr_method_pie(all_history_ko)
savefig('submintion_stat.png')
```

```python
dynamic_method_graph(ko, all_history_ko,'M')
savefig('ko_year_dynamic.png')
```

```python
dynamic_method_graph(ko, all_history_ko,'AS')
savefig('ko_month_dynamic.png')
```

```python
cmp_best_method = stats[stats.METHOD_DESC.isin([all_history_sub.index[0], all_history_ko.index[0]])]
best_methods = cmp_best_method.pivot_table('METHOD','DATE','METHOD_DESC',aggfunc='count').fillna(0)
```

```python
best_methods.resample('M','sum').dropna().plot(figsize=(15,5))
savefig('correlation_ko_sub.png')
```

```python
best_methods.resample('AS','sum').dropna().plot(figsize=(15,5))
savefig('correlation_month_ko_sub.png')
```
