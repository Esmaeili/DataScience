```python
from pandas import read_csv, DataFrame
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
```

```python
def distr_method_pie(history):
    plt.figure(figsize=(7,7))
    method_stats = history[history/float(history.sum()) > .02]
    colors=['yellow','green','red','blue', 'lightskyblue', 'lightcoral','gray', 'white']

    max_part = [0.1] + [0 for i in xrange(method_stats.__len__()-1)]

    plt.pie(method_stats, 
            labels=method_stats.index.tolist(), 
            explode=tuple(max_part),
            colors=colors,
            autopct='%1.1f%%', 
            shadow=True)
    plt.show()
```

```python
def dynamic_method_graph(method, history, diap):
    top_method = history[:4].index
    tmp = method[method.METHOD_DESC.isin(top_method)]
    top_method_grop = tmp.groupby(['DATE','METHOD_DESC']).count().METHOD.reset_index()
    pivot_method = top_method_grop.pivot('DATE','METHOD_DESC','METHOD').fillna(0)
    pivot_method.resample(diap,'sum').dropna().plot(figsize=(15,5))
```

```python
regr_analys_data = stats[stats.METHOD.isin(['SUB','KO/TKO'])]
```

```python
regr_analys_data.METHOD_DESC.fillna('Other', inplace=True)
```

```python
method_coder = LabelEncoder()
method_coder.fit(regr_analys_data.METHOD.drop_duplicates())
regr_analys_data['METHOD'] = method_coder.transform(regr_analys_data.METHOD)

method_desc_coder = LabelEncoder()
method_desc_coder.fit(regr_analys_data.METHOD_DESC.drop_duplicates())
regr_analys_data['METHOD_DESC'] = method_desc_coder.transform(regr_analys_data.METHOD_DESC)

all_fighters = regr_analys_data.FIGHTER_LOSE.append(regr_analys_data.FIGHTER_WIN)
all_fighters.drop_duplicates(inplace=True)
fighter_coder = LabelEncoder()
fighter_coder.fit(all_fighters)
regr_analys_data['FIGHTER_LOSE'] = fighter_coder.transform(regr_analys_data.FIGHTER_LOSE)
regr_analys_data['FIGHTER_WIN'] = fighter_coder.transform(regr_analys_data.FIGHTER_WIN)

DataFrame(method_desc_coder.classes_).to_csv('..\DataSets\ufc\method_description.csv', ';')
DataFrame(fighter_coder.classes_).to_csv('..\DataSets\ufc\dict_fighter.csv', ';')
```
