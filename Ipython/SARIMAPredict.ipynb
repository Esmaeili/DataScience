Теперь же загрузим, нужные библиотеки:

```python
from pandas import read_csv, DataFrame, Series
import statsmodels.api as sm
import rpy2.robjects as R
from rpy2.robjects.packages import importr
import pandas.rpy.common as com
from pandas import date_range
```

Теперь, как и в предыдущей статье, загрузим данные и перейдем к недельным интервалам:

```python
dataset = read_csv('DataSets/tovar_moving.csv',';', index_col=['date'], parse_dates=['date'], dayfirst=True)
otg = dataset.qty
w = otg.resample('w', how='sum')
w.plot(figsize=(12,6))
```

Итак, из графика можно заметить ежегодную сезонность (52 недели) и ярко выраженный тренд. Поэтому перед построением модели нам необходимо избавиться от тренда и сезонности.

#### Предварительный анализ данных

Итак для начала, прологорифмируем исходный ряд, для выравнивания значений: 

```python
w_log = log(w)
w_log.plot(figsize=(12,6))
```

Как видно у нас в графике присутствует сезонность и соответственно ряд не стационарен. Проверим это с помощью теста Дикки-Фулера, который проверяет гипотизу о наличии единичных корней и соотвтвенно если они есть ряд будет не стационарным. Как провести данный тест с помощью библиотеки <b>statsmodels</b>, я показывал в прошлый раз. Сейчас я продемонстрирую как это можно сделать с помощью функции <a href="http://www.inside-r.org/packages/cran/tseries/docs/adf.test">adf.test()</a> из <b>R</b>.

Итак, данная функция находится в R-ой библиотеке <a href="http://cran.r-project.org/web/packages/tseries/index.html">tseries</a>. Она предназначена для анализа временных рядов и устанавливается дополнительно. Загрузить нужную библиотеку можно с помощью функции <a href="http://rpy.sourceforge.net/rpy2/doc-dev/html/robjects_rpackages.html?highlight=importr#rpy2.robjects.packages.importr">importr()</a>.

```python
stats = importr('stats')
tseries = importr('tseries')
```

Можно заметить, что кроме <i>tseries</i>, мы загрузили еще и библиотеку <i>stats</i>. Она нам понадобитьсядля преобразования типов.

Теперь необходимо перевести данные из типа Python в тип понятный R. Сделать это можно с помощью функции <a href="http://pandas.pydata.org/pandas-docs/stable/r_interface.html">convert_to_r_dataframe()</a> на вход которой подается DataFrame, а на выходе получается вектор для R

```python
r_df = com.convert_to_r_dataframe(DataFrame(w_log))
```

Итак, вектор есть следующим шагом надо перевести его в формат временного ряда. Для этого в R существует функция <a href="http://stat.ethz.ch/R-manual/R-patched/library/stats/html/ts.html">ts()</a>, вызов ее будет выглядеть так:

```python
y = stats.ts(r_df)
```

Предварительная подготовка данных закончена и мы можем вызвать нужную нам функцию:

```python
ad = tseries.adf_test(y, alternative="stationary", k=52)
```

В качестве параметров ей передается временный ряд и кол-во лагов, для которых будет расчитываться тест. В экономических моделях принято брать данное значение равное году, а т.к. данные у нас еженедельные, а в году 52 недели, поэтому параметр имеет такое значение.

Теперь в переменной <i>ad</i> содержится R-объект. Его структура описана в виде списка, описание которого мне найти не удалось. Поэтому с помощью визуального анализа я написал код, который выводит результат работы функции в понятном виде:

```python
a = ad.names[:5]
{ad.names[i]:ad[i][0] for i in xrange(len(a))}
```

Исходя из результатов теста, на исходный ряд не стационарен. Соответственно теперь проверим на стационарность ряд первых разностей.

Для начала получим первые разности с помощью Python, а затем применим ADF-тест:

```python
diff1lev = w.diff(periods=1).dropna()
print 'p.value: %f' % sm.tsa.adfuller(diff1lev, maxlag=52)[1]
```

```python
diff1lev.plot(figsize=(12,6))
```

Ряд первых разностей, по итогам теста, оказался стационарным. А график помогает нам убедиться, что тенденция отсутствует. Осталось избавиться от сезонности.

Для этого нужно взять сезонную разность, от нашего получившегося ряда. Подробнее можно прочитать <a href="https://onlinecourses.science.psu.edu/stat510/?q=node/67">тут</a>. Если полученный ряд будет стационармы необходимо будет вязть его превую разность и проверить ее.

Возьмем сезонную разность:

```python
diff1lev_season = diff1lev.diff(52).dropna()
```

Проверим ее на стационарность с помощью ADF-тест из R:

```python
r_df = com.convert_to_r_dataframe(DataFrame(diff1lev_season))
y = stats.ts(r_df)
ad = tseries.adf_test(y, alternative="stationary", k=52)
a = ad.names[:5]
{ad.names[i]:ad[i][0] for i in xrange(len(a))}
```

Итак, ряд не стационарен, возьмем его первые разности:

```python
diff1lev_season1lev = diff1lev_season.diff().dropna()
print 'p.value: %f' % sm.tsa.adfuller(diff1lev_season1lev, maxlag=52)[1]
```

Получившийся ряд стационарен. Теперь можно перейти к построению модели 

#### Построение модели.

Итак, предварительный анализ закончен, и мы можем перейти к построению <i>сезонной модели ARIMA</i> <b>(SARIMA)</b>.

Общий вид данной модели $ARIMA(p,d,q)(P,D,Q)_s$

В этой модели параметры обозначают следующее:

<ul>
<li>$p$ - порядок модели $AR(p)$</li>
<li>$d$ - порядок интегрирования исходных данных</li>
<li>$q$ - порядок модели $MA(q)$</li>
<li>$P$ - порядок сезонной составляющей $SAR(P)$</li>
<li>$D$ - порядок интегрирования сезонной составляющей</li>
<li>$Q$ - порядок сезонной составляющей $SMA(Q)$</li>
<li>$s$ - размерность сезонности(месяц, квартал и т.д.)</li>
</ul>

Как определять <b>p</b>, <b>d</b>, <b>q</b>, я показывал в прошлый раз. Сейчас я опишу определять порядок сезонных составляющих <b>P</b>,<b>D</b>,<b>Q</b>. 

Начнем с определения параметра <b>D</b>. Он определет порядок интегрированности сезонной разности, т.е. в нашем случае он равен <b>1</b>. Для определения <b>P</b> и <b>Q</b> нам как и прежде надо построить коррелограммы <i>ACF</i> и <i>PACF</i>  

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff1lev_season1lev.values.squeeze(), lags=150, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff1lev_season1lev, lags=150, ax=ax2)
```

Из гарфика <i>PACF</i> видно, что порядок AR будет <b>p=4</b>, а по <i>ACF</i> видно, что порядок MA <b>q = 13</b>, т.к. 13 лаг - это последний лаг отличный от 0.  

Теперь перейдем к сезонным составляющим. Для их оценки надо смотреть на лаги кратные размеру сезонности, т.е., если для нашего примера, сезонность 52, то надо рассматривать лаги 52, 104, 156, ...

В нашем случае параметры <b>P</b> и <b>Q</b> будут равны 0 (это видно если посмотреть на ACF и PACF на указанных выше лагах).

В результате наших исследований мы получили модель $SARIMA(4,1,13)(0,1,0)_{52}$

Как было указано в начале данной статьи, что найти способы построения данной модели на Python я не нашел, поэтому я принял решение воспользоваться для это функцией <a href="http://stat.ethz.ch/R-manual/R-patched/library/stats/html/arima.html">arima()</a> из R. В качестве параметров ей передаются порядок модели ARIMA и, при необходимости, порядок сезонной составляющей. Но перед вызовом ее необходимо подготовить некоторые данные.

Для начала переведем наш исходный набор в формат R и переведем в формат временного ряда.

```python
r_df = com.convert_to_r_dataframe(DataFrame(w))
y = stats.ts(r_df)
```

Порядок модели передается в качестве вектора R, поэтому давайте создадим его:

```python
order = R.IntVector((4,1,13))
```

Так же в качестве параметра сезонной составляющей передается список, который содержит ее порядок и размер периода:

```python
season = R.ListVector({'order': R.IntVector((0,1,0)), 'period' : 52})
```

Теперь мы готовы к тому, чтобы построить модель:

```python
model = stats.arima(y, order = order, seasonal=season)
```

Итак наша модель готова и можно перейти к построению прогноза на ее основе.

#### Проверка адекватности модели

Итак для проверки адекватности модели надо проверить соответвуют ли остатки модели "белому шуму". Проверим это проведя Q тест и проверим корреляцию остатков. Для этого в R существует функция <a href="">tsdiag()</a>, в качестве параметра передается модель и количество лагов для теста.

Вызвать данную функцию можно так: 

```python
%Rpush model
%R tsdiag(model, 100)
```

В первой строке инструкция <b><i>%Rpush</i></b> загружает объекты для использования в R. Иструкция <b><i>%R</i></b> во второй строке вызывает код в формате языка R. Данная конструкция работает в <b>IPython Notebook</b>.

Из графиков выше можно заметить, что остатки независимы (это видно по <i>ACF</i>). Кроме того из графика Q-статистики можно заметить что во всех точках значение p-value больше уровня значимости, из этого можно сделать вывод, что остатки, с большой вероятностью, являются "белым шумом".

#### Прогнозирование

Для прогонизования нужно подгрузить библиотеку <a href="http://cran.r-project.org/web/packages/forecast/index.html">forecast</a>

```python
forecast = importr('forecast')
```

Вывести результы прогнозирования можно двумя способами.

<b>Способ 1.</b> Это использовать возможности интеграции IPython и R. Который был показан в предыдущем разделе:

```python
%R plot(forecast(model))
```

<b>Способ 2.</b> Второй способ это сделать прогноз с помощью библиотеки <i>forecast</i>, а потом перевести результат в во временную серию pandas и вывести их на экран. Код который будет выполнять написан ниже:

```python
f = forecast.forecast(model) #строим прогноз
pred = [i[1] for i in f[3].iteritems()] #парсим результат
dt = date_range(w.index[-1], periods=len(pred)+1,freq='W')[1:] #создаем индекс из дат
pr = Series(pred, index = dt)  #записываем данные в TimeSeries
```

Теперь выведем результат на экран

```python
w.plot(figsize=(12,6))
pr.plot(color = 'red')
```
